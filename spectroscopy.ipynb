```python
import numpy as np
import matplotlib.pyplot as plt

# Read data from file 2 columns
def read_data(file_name):
    x = []
    y = []
    with open(file_name, 'r') as file:
        next(file)  # skip header
        for line in file:
            values = line.strip().split(',')
            x.append(float(values[0]))
            y.append(float(values[1]))
    return np.array(x), np.array(y)

# Linear regression
def linear_regression(x, y):
    n = len(x)
    x_mean = np.mean(x)
    y_mean = np.mean(y)
    
    #coefficients
    numerator = np.sum((x - x_mean) * (y - y_mean))
    denominator = np.sum((x - x_mean) ** 2)
    
    slope = numerator / denominator
    intercept = y_mean - slope * x_mean
    
    return slope, intercept

# Linear function
def predict(x, slope, intercept):
    return slope * x + intercept

# Main
if __name__ == "__main__":
    # Read data from data.txt
    x, y = read_data('data.txt')

    # Calculate
    slope, intercept = linear_regression(x, y)

    print(f'Slope coefficient: {slope}')
    print(f'Free term: {intercept}')
  # Linear graph
    y_pred = predict(x, slope, intercept)

    # Plot
    plt.scatter(x, y, color='blue', label='Experimental data')
    plt.plot(x, y_pred, color='red', label='Regression')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Linear regrassion')
    plt.legend()
    plt.show()

```
